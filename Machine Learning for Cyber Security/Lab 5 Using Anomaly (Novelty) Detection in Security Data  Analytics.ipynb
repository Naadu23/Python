```python
#Import libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import OneClassSVM
from sklearn.metrics import confusion_matrix, classification_report
```

```python
#Load the benign dataset
bdata=pd.read_csv('Benign.csv', header=None)

bdata
```

```python
#Load the attack dataset
adata=pd.read_csv('Attack.csv', header=None)
adata
```

```python
#Remove missing values
bdata.dropna(inplace=True)
adata.dropna(inplace=True)

#label benign as True
bdata['Type'] = True
```

###   Step 4: Split the benign dataset into 80:20 for training and testing

```python
#split benign
trainSetB, testSetB = train_test_split(bdata, test_size=0.2, random_state=42)
```

###   Step 5: Fit a One-Class Support Vector Machine (One-Class SVM)

```python
#separate features and labels
x_train = trainSetB.drop(columns='Type')
y_train = trainSetB['Type']

#Train model
model = OneClassSVM(kernel='rbf', gamma='auto').fit(x_train)

model
```

###   Step 8 & 9: Add labels to the attack data (label as False)

```python
# Step : Add labels to the attack data (label as False)
adata['Type'] = False

#Step 9: Pool the benign test instances with malicious instances
testSet = pd.concat([testSetB, adata], ignore_index=True)
```

###   Step 10: Model evaluation (OCSVM)

```python
#separate features and labels for testing
x_test = testSet.drop(columns='Type')
y_test = testSet['Type']
```

```python
#Make prediction
pred = model.predict(x_test)

#convert predictions to true (normal) and false(malicious)
pred =[True if p ==1 else False for p in pred]

#Evaluate
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```
