```python
#Import necessary libraries
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load iris dataset
iris = load_iris()
iris
```

```python
#convert the dataset to a dataframe
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df
```

```python
#initialize the KMeans model with 3 clusters
kmeans = KMeans(n_clusters=3,random_state=42)

#fit model to data
kmeans.fit(df)
```

```python
#predict the clusters
predicted_clusters = kmeans.predict(df)

#add predicted clusters to df
df['Cluster'] = predicted_clusters

#display df
df
```

```python
#visualize clusters (for simplicity using first two features)
plt.scatter(df.iloc[:,0], df.iloc[:,1], c=predicted_clusters, cmap='viridis', marker='o',edgecolor='k')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title('K-Means Clustering of Iris Dataset')
plt.show()
```

###   Task 2 : Simple Linear Regression

```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

#create a dataset x for hours, y for exams score
x = np.array([[1], [2], [3], [4], [5]])
y = np.array([1,2,3,3.5,5])
```

```python
#initislize model
model = LinearRegression()

#fit model
model.fit(x,y)

#predict y based on model
y_pred = model.predict(x)

# Show slope
print("Slope (coefficient):", model.coef_[0])
print("Intercept:", model.intercept_)
```

```python
#plot original data
plt.scatter(x,y, color='blue',label='Original data')

#plot regression line
plt.plot(x,y_pred, color='red', label='Regression line')

#label axes and plot
plt.xlabel('Independent Variable (x)')
plt.ylabel('Dependent Variable (y)')
plt.title('Simple Linear Regression')

#show legend
plt.legend()

#show plot
plt.show()
```

###   Task 3 : Network Traffic Analysis Using Clustering Algorithms

```python
#load dataset
network_data = pd.read_csv('UNSW_NB15_training-set.csv')
network_data
```

```python
network_data.columns
```

```python
#drop labels
net_data1 = network_data.drop(columns=['label','attack_cat'])
```

```python
#drop categorical data
net_data = net_data1.drop(columns=['id','proto','service','state'])
```

```python
#normalize data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
net_data_normalized = scaler.fit_transform(net_data)
```

```python
from sklearn.cluster import KMeans

kmeans_net = KMeans(n_clusters=4, random_state=42)
kmeans_net.fit(net_data_normalized)

#interpret results
clusters = kmeans_net.labels_
print(f"Cluster centers: {kmeans_net.cluster_centers_}")
```
